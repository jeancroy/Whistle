```python
#Todo compute Help Bonus(berries,ing) frpm the modifiers
# item = species, obeservation = both individual and recording ?

def computeBonusAndSpeed(individual):
    
    species = pokedex_dict[individual['Name']]
    berryBonus = 2.0 if species["Specialty"] == 'Berries' else 1.0
    ingBonus = 2.0 if species["Specialty"] == 'Ingredients' else 1.0
    
    levelBonus = (1 + (individual["Level"]-1)*2.0/1000)
    help3h = species["Helps_3h"]*levelBonus
    
    return (help3h, ingBonus, berryBonus)

def berryCount(help3h, berryBonus, ingredient_fraction, energy_m):
    return int(round(help3h*(1-ingredient_fraction)*energy_m*berryBonus))

def ingCount(help3h, ingBonus, ingredient_fraction, energy_m):
    return int(round(help3h*ingredient_fraction*energy_m*ingBonus))

def validate(observation, computedBonusAndSpeed, ingredient_fraction, energy_m):
    
    help3h, ingBonus, berryBonus = computedBonusAndSpeed
    
    ing = berryCount(help3h, ingBonus, ingredient_fraction, energy_m)
    ber = ingCount(help3h, berryBonus, ingredient_fraction, energy_m)

    return ing == int(observation["Ingredients"]) and ber == int(observation["Berries"])
```
